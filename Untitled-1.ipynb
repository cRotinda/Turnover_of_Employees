```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as matplot
import seaborn as sns
import os
%matplotlib inline
```

```python
##Importing the csv file into Pandas
df=pd.read_csv("C:\Projects\Employee_Turnover/turnover.csv")
df
```

```python
##Checking for Null Values
df.isnull().any()
```

```python
##Renaming the columns for easier acccess
df = df.rename(columns={
    "satisfaction_level": "satisfaction",
    "last_evaluation": "evaluation",
    "number_project": "projectCount",
    "average_montly_hours": "averageMonthlyHours",
    "time_spend_company": "yearsAtCompany",
    "Work_accident": "workAccident",
    "left": "turnover",
    "promotion_last_5years": "promotion",
    "sales": "department",
    "salary": "pay"
})
df
```

```python
##Putting the hasLeft Column first for better understanding
front=df["turnover"]
df.drop(labels="turnover",axis=1,inplace=True)
df.insert(0,"turnover",front)
df.head()
```

```python
##The dataset has 14999 rows and 10 columns
df.shape
```

```python
##Showing the precentage of the employees who stayed and left
turnover_value=df.turnover.value_counts()/len(df)
turnover_value
```

```python
turnover_Summary=df.groupby("turnover")
turnover_Summary.mean()
```

```python
empl_satisfaction=df["satisfaction"][df["turnover"]==0].mean()
turnover_satisfaction=df["satisfaction"][df["turnover"]==1].mean()
print("The average satisfaction of a working employee is "+str(empl_satisfaction))
print("The average satisfaction of a leaving employee is "+str(turnover_satisfaction))
```
